### Homework Explanation

The question asks to take input from the user using the `input()` function to accept a Celsius temperature and convert it to Fahrenheit.

This is done using the equation: $F = (C \times \frac{9}{5}) + 32$

```python
c = float(input("Enter a Celsius temperature: "))
f = (c * (9/5))+32 
print(f"{c} Celsius is equal to {round(f,2)} Fahrenheit.")
```
